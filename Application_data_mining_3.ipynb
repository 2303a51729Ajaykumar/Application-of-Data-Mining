```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/patients.csv')
```

```python
df['Glucose_Level'].fillna(df['Glucose_Level'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Blood_Pressure'].fillna(df['Blood_Pressure'].mean(),inplace=True)
```

```python
df['Insulin'].fillna(df['Insulin'].mean(),inplace=True)
```

```python
df['BMI'].fillna(df['BMI'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python


  scaler = MinMaxScaler()
```

```python
df[['Age', 'Glucose_Level']] = scaler.fit_transform(df[['Age', 'Glucose_Level']])
```

```python
df['Blood_Pressure'] = (df['Blood_Pressure'] - df['Blood_Pressure'].min()) / (df['Blood_Pressure'].max() - df['Blood_Pressure'].min())
```

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python
df[['BMI']] = s_scaler.fit_transform(df[[ 'BMI']])
```

```python
df['BMI']
```

```python
df['Insulin'] = (df['Insulin'] - df['Insulin'].mean()) / df['Insulin'].std()
```
